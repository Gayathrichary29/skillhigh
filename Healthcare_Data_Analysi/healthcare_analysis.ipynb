```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("healthcare_data.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
disease_count = df['Disease'].value_counts()
```

```python
plt.figure(figsize=(8,5))
disease_count.plot(kind='bar')
plt.title("Disease Distribution")
plt.xlabel("Disease")
plt.ylabel("Number of Patients")
plt.show()
```

```python
age_disease = df.groupby('Disease')['Age'].mean()
```

```python
plt.figure(figsize=(8,5))
age_disease.plot(kind='bar')
plt.title("Average Age by Disease")
plt.xlabel("Disease")
plt.ylabel("Average Age")
plt.show()
```

```python
gender_disease = df.groupby(['Disease', 'Gender']).size().unstack()
```

```python
gender_disease.plot(kind='bar')
plt.title("Gender-wise Disease Distribution")
plt.xlabel("Disease")
plt.ylabel("Patient Count")
plt.show()
```

```python
bp_disease = df.groupby('Disease')['BloodPressure'].mean()
```

```python
plt.figure(figsize=(8,5))
bp_disease.plot(kind='bar')
plt.title("Average Blood Pressure by Disease")
plt.xlabel("Disease")
plt.ylabel("Blood Pressure")
plt.show()
```

```python
diabetes_count = df['Diabetes'].value_counts()
```

```python
plt.figure(figsize=(6,4))
diabetes_count.plot(kind='bar')
plt.title("Diabetes Prevalence")
plt.xlabel("Diabetes")
plt.ylabel("Count")
plt.show()
```
